```python
import pandas as pd
```

```python
data=pd.read_csv('clean_data.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
x=data[['ph','Hardness','Solids','Chloramines','Sulfate','Conductivity','Organic_carbon','Trihalomethanes','Turbidity']]
y=data['Potability']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=0)
```

```python
x_test
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
lg= KNeighborsClassifier(n_neighbors=19, metric='minkowski', p=2)
```

```python
lg.fit(x_train,y_train)
```

```python
pridict=lg.predict(x_test)
```

```python
pridict
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac=accuracy_score(y_test,pridict)
```
